```python
# 初始化 Binarizer
binarizer = Binarizer(threshold=0.0)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
binarized_data = binarizer.fit_transform(numeric_df)
binary_df=df.copy()
binary_df[numeric_df.columns] = binarized_data
binary_df
```

```python
# 初始化OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# 對"position"欄位進行one-hot encoding
encoded_positions = encoder.fit_transform(df[['Position']])

# 取得One-Hot Encoding之後的欄位名稱
encoded_columns = encoder.get_feature_names_out(['Position'])

# 將結果轉換為DataFrame並與原始df合併
encoded_df = pd.DataFrame(encoded_positions, columns=encoded_columns)

# 合併原始df與編碼後的df
onehot_df = pd.concat([df, encoded_df], axis=1)
onehot_df
```

```python
# 創建 KBinsDiscretizer 實例，設置 K 值和策略
k_bins = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
discrete_df=df.copy()
# 選擇所有數值型欄位
numeric_cols = discrete_df.select_dtypes(include=['float64', 'int64']).columns

# 對所有數值型欄位進行離散化
for col in numeric_cols:
    binned_values = k_bins.fit_transform(discrete_df[[col]])
#     discrete_df[col + '_Binned'] = binned_values.flatten()
discrete_df
```

```python
# 創建 StandardScaler 實例
scaler = StandardScaler()

# 選擇所有數值型欄位
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
stander_df=df.copy()
# 對所有數值型欄位進行標準化
stander_df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
stander_df
```

```python
scale_df=df.copy()

numeric_cols = scale_df.select_dtypes(include=['float64', 'int64']).columns
scale_df[numeric_cols] = scale(scale_df[numeric_cols])
scale_df
```

```python
minmax_df=df.copy()
# 創建 MinMaxScaler 實例
scaler = MinMaxScaler()

# 選擇所有數值型欄位
numeric_cols = minmax_df.select_dtypes(include=['float64', 'int64']).columns

# 對所有數值型欄位進行縮放
minmax_df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
minmax_df
```

```python
pca_df=stander_df.copy()

# 使用 select_dtypes() 選擇數值型欄位
numeric_cols = pca_df.select_dtypes(include=['float64', 'int64']).columns

# 對數值型欄位進行標準化
scaler = StandardScaler()
df_scaled = scaler.fit_transform(pca_df[numeric_cols])

# 創建 PCA 實例，設置要保留的主成分數量
pca = PCA(n_components=3)

# 對標準化後的數據進行 PCA
principal_components = pca.fit_transform(df_scaled)

# 將主成分轉換為 DataFrame
principal_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2', 'Principal Component 3'])
principal_df
```

```python
Var_df=df.copy()
# 使用 select_dtypes() 選擇數值型欄位
numeric_cols = Var_df.select_dtypes(include=['float64', 'int64']).columns

# 設置變異量閾值，這裡設為 0.1
threshold = 0.1
selector = VarianceThreshold(threshold=threshold)

# 擬合並轉換數據，只對數值型欄位進行操作
df_filtered = selector.fit_transform(Var_df[numeric_cols])

# 將篩選後的數據轉換為 DataFrame
df_filtered = pd.DataFrame(df_filtered, columns=numeric_cols[selector.get_support()])
df_filtered
```

```python
best_df=df.copy()

# 分割數據和目標變數
X = best_df.select_dtypes(include=['float64', 'int64']).drop(columns='Attack')  # 選擇數值型欄位
y = best_df['Attack']  # 目標變數

# 使用 SelectKBest 選擇最好的 K 個特徵
k = 3  # 設置要選擇的特徵數量
selector = SelectKBest(score_func=f_classif, k=k)

# 擬合選擇器並轉換數據
X_selected = selector.fit_transform(X, y)

# 獲取選擇的特徵的名稱
selected_features = X.columns[selector.get_support()]

# 將選擇的特徵轉換為 DataFrame
X_selected_df = pd.DataFrame(X_selected, columns=selected_features)

# 將選擇的特徵與目標變數合併
result_df = pd.concat([X_selected_df, y.reset_index(drop=True)], axis=1)
result_df
```
